```python
df=pd.read_csv("world-happiness-report-2021.csv")
df
```

```python
# Lista de nombres de columnas a eliminar
columnas_a_eliminar = ["upperwhisker", "lowerwhisker","Explained by: Log GDP per capita", "Explained by: Social support", 
                       "Explained by: Healthy life expectancy", "Explained by: Freedom to make life choices","Explained by: Generosity",
                       "Explained by: Perceptions of corruption","Dystopia + residual"]

# Eliminar las columnas utilizando el método .drop()
df = df.drop(columnas_a_eliminar, axis=1)

# axis=1 indica que estamos eliminando columnas (axis=0 sería para filas)

# Si deseas hacer cambios en el DataFrame original, puedes usar el argumento inplace=True
# df.drop(columnas_a_eliminar, axis=1, inplace=True)
```

```python
# Crear un diccionario con los nuevos nombres de las columnas
nuevos_nombres = {"Country name":"Country"}

# Cambiar los nombres de las columnas utilizando el método .rename()
df = df.rename(columns=nuevos_nombres)
```

```python
df
```

```python
def data_exploring(csv):
    forma = csv.shape
    print(f"La forma es {forma}")
    print("_______________")
    columnas = csv.columns
    print(f"Las columnas son {columnas}")
    print("_______________")
    nulos = csv.isna().sum().reset_index()
    print(f"Los nulos son:")
    display(nulos)
    print("_______________")
    duplicados = csv.duplicated().sum()
    print(f"Hay {duplicados} duplicados")
    print("_______________")
    tipo_dato = csv.dtypes.reset_index()
    print(f"Los datos son de tipo:")
    display(tipo_dato)
```

```python
data_exploring(df)
```
